```python
#8 سوال سكشن الافلاين 
students=[]
def add_students():
    name=input("enter student name")
    marks={}
    subjects=["math","science","english"]
    for sub in subjects:
        mark=float(input(f'enter {sub} mark'))
        marks[sub]=mark
    student ={
        "name":name,
        "marks":marks
    }
    
    students.append(student)
def show_students():
    if not students:
        print("not found")
        return
    for s in students:
        marks=s["marks"]
        avg=sum(marks.values())/len(marks) 
        status="pass"if avg>=50 else "fail"
            
        print("name",s["name"])
        print("marks",marks)
        print("average",avg)
        print("status",status)
        print("-"*20)
def search_student():
    name=input("enter student name")
    for s in students:
        if s["name"].lower()==name.lower():
            print("student found")
            print("name",s["name"])
            print("marks",s["marks"])
            return
    print("student not found")
def delete_student():
    name=input("enter student name")
    for s in students:
        if s["name"].lowr ()==name.lower():
            students.remove(s)
            print("student deleted successfully")
            return
    print("student not found")
def menu():
    while True:
        print("1.add student")
        print("2.show student")
        print("3.search student")
        print("4.delete student")
        print("5.exit")
        
        choice=input("enter your choice: ")
        if choice=="1":
            add_students()
        elif choice=="2":
            show_students()
        elif choice=="3":
            search_student()
        elif choice=="4":
            delete_student()
        elif choice=="5":
            print("goodbye")
            break
        else:
            print("invaid choice,try again")
menu()   
```
