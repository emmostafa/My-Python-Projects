```python
#take a string from the user and print a number of vowels (الحروف المتحركه) and consonants (صوت الحرف )
def count_letters():
    text = input("Enter a string: ").lower()
    vowels = "aeiou"
    v_count = 0
    c_count = 0

    for i in text:
        if i.isalpha():#  علشان نتاكد انه حرف وليس رقم 
            if i in vowels:
                v_count += 1
            else:
                c_count += 1

    print("Number of vowels:", v_count)
    print("Number of consonants:", c_count)

count_letters()
```

```python
#Write a program that asks the user for numbers until they type 'stop', then prints the sum and avarage
total = 0
count = 0

while True:
    num = input("Enter a number (or type 'stop' to finish): ")
    
    if num == "stop":
        break
    
    num = float(num)
    total += num
    count += 1

if count > 0:
    average = total / count
    print("Sum =", total)
    print("Average =", average)
else:
    print("No numbers were entered.")
```

```python
#15) Write a function that takes a list and returns a new list containing only prime numbers
def i_prime(num):
    if num<=1:
        return False
    for i in range(2,int(num**.5)+1):
        if num%i==0:
            return False
    return True
def x_primes(numbers):
    list5=[]
    for num in numbers:
        if i_prime(num):
            list5.append(num)
    return list5
print(x_primes([25,66,34,22,21,5,3,2]))
```
