```python
#max,min,avarg
def start_val(data):
    total=0
    mx=data[0]
    mn=data[0]
    for i in data:
        total+=i
        if i>mx:
            mx=i
        if i<mn:
            mn=i
        avarg=total/len(data)
        return {"max",mx,"min",mn,"avarage",avarg}
print(start_val([4,5,6,7]))
```
