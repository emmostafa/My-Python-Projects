```python
#برنامج يعد عدد الكلمات وحصر عدد مرات التكرار لكل كلمه فى القاموس
dic={}
x=input('enter the text')
words=x.split()
for word in words:
    if word in dic:
        dic[word]+=1
    else:
        dic[word]=1
print(dic)
    
```

```python
#قايمه بها مجموعه كلمات عايزه اضع الا ليها نفس الحجم فى قايمه 
names =['eman','osama','eslam','sama']
group={}
for word in names:
    l=len(word)
    if l not in group:
        group[l]=[]
    group[l].append(word)
for l,word in group.items():
    print(f'حروف {l}كلمات من :{word}')
print(group)
        
```

```python
#Write a program that:
#Stores items and prices using a dictionary and Uses a loop to let the user: Add items,Stop when finished,Calculates total cost
dic={'apple':200,'egg':100,'banana':300}
total=0
while True:
    key=input()
    if key=='excit':
        break
    if key in dic:
        price=dic[key]
        total+=price
    else:
        print('هذا المنتج مش موجود')
print(f'المجموع الكلى {total}')    
        
```
