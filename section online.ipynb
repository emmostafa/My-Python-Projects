```python
#برنامج يعد عدد الكلمات وحصر عدد مرات التكرار لكل كلمه فى القاموس
dic={}
x=input('enter the text')
words=x.split()
for word in words:
    if word in dic:
        dic[word]+=1
    else:
        dic[word]=1
print(dic)
    
```

```python
#اليوزر يدخل نص والبرنامج يشيل كل الحروف المكرره 
text= input ('enter your text')
x=''
for i in text:
    if i not in x:
        x+=i
print(x)
```

```python
#قايمه بها مجموعه ارقام الكود يطلع تانى اكبر عدد فى القايمه
l1=[1,3,4,5,6,78,33,22]
maxnum=l1[0]
secnum=l1[0]
for i in l1:
    if i >maxnum:
        secnum=maxnum
        maxnum=i
print(maxnum)
print(secnum)
```

```python
#اليوزر هيدخل نص عايز النص يتطبع بطريقه معكوسه
text= input()
word=text.split()
x=len(word)-1
while x >0:
    print(word[x],end=' ')
    x-=1
```

```python
#قايمه بها مجموعه كلمات عايزه اضع الا ليها نفس الحجم فى قايمه 
names =['eman','osama','eslam','sama']
group={}
for word in names:
    l=len(word)
    if l not in group:
        group[l]=[]
    group[l].append(word)
for l,word in group.items():
    print(f'حروف {l}كلمات من :{word}')
print(group)
        
```
