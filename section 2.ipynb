```python
#برنامج لحساب اصغر رقم بدوان استخدام الداله max
smallest=None
while True:
    user_input=input("enter a number(done or finish)")
    if user_input=="done":
        break
    number=float(user_input)
    if smallest is None:
        smallest = number
    elif number < smallest:
        smallest=number
if smallest is not None:
    print("the smallest number if :" , smallest)
else:
    print("no numbers were entered")
```

```python
#برنامج تحويل القيم الى مفاتيح والمفاتيح الى قيم 
data={"A":1,"B":2,"c":1}
inverted={}
for key,value in data.items():
    if value not in inverted:
        inverted[value]=[]
    inverted[value].append(key)
print(inverted)
```

```python
students=[
    {"name":"eman","grades":[50,49,50]},
    {"name":"safa","grades":[40,49,50]},
    {"name":"shahd","grades":[50,59,44]}
]
best_student=""
high_ave=0
for s in students:
    name=s["name"]
    grades=s["grades"]
    ava=sum(grades)/len(grades)
    print(f"{name} is avarage:{ava}")

    if ava>high_ave:
       high_ave=ava
       best_student=name
    
print(f"top student {best_student} with avarage{high_ave}")
    
```
