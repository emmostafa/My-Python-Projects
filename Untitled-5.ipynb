```python
#برنامج ايجاد تانى اكبر رقم فى القايمه
list=[2,3,4,5]
unique_sorted=sorted(set(list))#لازاله الارقام المكرره
result=unique_sorted[-2]#دا تانى اكبر عنصر الاكبر الاخر
print(result)
#برنامج لحساب عدد مرات حساب العد لو العنصر موجود نزود العداد 1
list=["Eman","Eslam","fauroza","sama","fauroza","Eman"]
count={}
for item in list:
    if item in count:
        count[item]+=1
    else:
        count[item]=1
print(count)
print("-"*10)

#برنامج ازاله الاعداد السالبه من قايمه
list1=[]
list2=[2,3,4,-1,-2,-3]
for i in list2:
    if i>=0:
        list1.append(i)
print(list1)
print("-"*10)

#برنامج طباعه الاسامى الا تحتوى على حرف a وعدد حروفها اكبر من 4
x=[]
list=["Eman","mona","aya","yasmeen","ibraheem"]
for i in list:
    if "a" in i and len(i)>4:
        x.append(i)
print(x)


```

```python
#ادخال معاملات الخط المستقيم الاول نفرض مثلا a,b,c
a=float(input("enter a"))
b=float(input("enter b"))
c=float(input("enter c"))
#معاملات الخط المستقيم التانى نفرض مثلا E,m,n
E=float(input("enter E"))
v=float(input("enter m"))
n=float(input("enter n"))
#نجيب ميل كل خط مستقيم
m1=-a/b
m2=-E/v
#نشوف بقا هل المسيمان متعامده ام متوازيه ام متقاطعه ام متطابقه
if (m1*m2)==-1:
    print("الخطان متعامدان")
elif(m1==m2):#هنا فيه حالتين اما متوازيين اما متطابقين على حسب الجزء المقطوع لومتساوى يكون متطابقين غيز كدا متوزايين
    if (a/E == b/v ==c/n):
        print("متطابقان")
    else:
        print("متوازيان")
else:
    print("متقاطعان")
```
