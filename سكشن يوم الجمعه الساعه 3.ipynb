```python
#كود بيعمل عمليات ال atm
nam_owner=(input("enter u name"))
#رصيدك
balance=float(input("enter u balance"))
#ايداع مبلغ
eda3=float(input("enter the amount"))
balance+=eda3
print(f"تم الايداع {eda3}pount رصيدك الحالى{balance}")
#سحب مبلغ من رصيدك
bull=float(input("How many?"))
if bull<=balance:
    balance-=bull
    print(f"تم سحب{bull},رصيدك الحالى {balance}")
else:
    print("رصيدك غير كافى")
#الباقى
print(f"الرصيد الموجود{nam_owner} {balance}")
```

```python
#جدول بيه صفوف واعمده لعمل الشكل التالى
'''xox
oxo 
x0x'''
rows=int(input("please insert the number of rows:"))#ادخال عدد الصفوف
cols= int(input("please insert the number of cols:"))#ادخال عدد الاعمده
for i in range (rows):
    row= ""
    for j in range (cols):
        if (i + j)% 2==0:
            row += " x "
        else:
            row+=" 0 "
    print(row)
```

```python
#قايمه فيها ازواج مرتبه ونحط الا مجموعهم 10 فى قايمه
pairs =[(6,4),(2,4),(7,3),(5,5)]
degree_sum=10
list_target = []
for pair in pairs:
    if sum(pair) == degree_sum:
        list_target.append(pair)
print(list_target)
```
