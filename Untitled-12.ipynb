```python
def say_hello(skill1,skill2,skill3,skill4):
    skills=[skill1,skill2,skill3,skill4]
    for skill in skills:
        print(skill)
say_hello("Html","c#","js","python")
#say_hello("html","python","js","++c","c#") r:error
def show_details(*skills):
    for s in skills:
        print(s)
show_details("js","python","c++")
```

```python
def say_Hello(name,age,country="known"):
    print(f"hello {name} your age is {age} and country is {country}")
say_Hello("eman",19 )

def show_skills(**skills):
    for skill,value in skills.items():
        print(f"{skill} :{value}")
show_skills(html="60%",python="70%",js = "80%")
```

```python
#test word [wwwoooorrrld]
#x="wwwoooorrld"     print(x[1:])

def cleanword(word):
    if len(word)==1:
        return word
    print(f"print start function {word}")
    if word[0]==word[1]:
        print(f"print before condition {word}")
        return cleanword(word[1:])
    print(f"print before return {word}")
    return word[0]+cleanword(word[1:])
    #stash[word]
print(cleanword("wwwoooorrld"))
```
