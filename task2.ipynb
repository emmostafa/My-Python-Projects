```python
#Write a program to look at a list of whole numbers and tell you how many of them are even
numbers = [11,3,5,6,8,10,4]
count = 0

for n in numbers:
    if n % 2 == 0:  # التحقق إذا كان الرقم يقبل القسمة على 2 بدون باقي
        count += 1

print(f"عدد الأرقام الزوجية هو: {count}")
```

```python
#Write a program to find the largest number in a list without using the built-in max() function
nums = [200,44,66,75,80]
max_val = nums[0]  # نفترض أن أول رقم هو الأكبر مبدئياً

for n in nums:
    if n > max_val:
        max_val = n  # تحديث القيمة الأكبر إذا وجدنا ما هو أكبر منها

print(f"أكبر رقم هو: {max_val}")
```

```python
#Write a program to read text and return a dictionary of word frequencies
text = "apple banana apple cherry banana apple"
words = text.split()
frequency = {}

for w in words:
    # إذا كانت الكلمة موجودة زد عدادها، وإلا ابدأ من 0
    frequency[w] = frequency.get(w, 0) + 1

print(frequency)
```

```python
#write a program to calculate max, min and average 
def get_statue(data):
    total=0
    max=data[0]
    min=data[0]
    for i in data:
        total+=i
        if i > max:max=i
        if i < min:min=i
    return {"max":max , "min":min , "aver":total/len(data)}
print(get_statue([3,4,7,8,9]))
            
```

```python
#Write a program to find the student with the highest average in a nested dictionary
students={
    "eman":{"math":60,"science":55},
    "yaso":{"math":70,"science":54}
}
max_student=""
max_avg=0
for name , grades in students.items():
    avg=sum(grades.values())/len(grades)
    if avg> max_avg:
        max_avg= avg
        max_student=name
print(f"top grade is :{max_student} ={max_avg}")
```
