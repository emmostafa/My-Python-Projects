```python
# 1. دالة تحميل البيانات من الملف
def load_hotel_data(filename):
    hotel_dict = {}
    try:
        with open(filename, 'r') as file:
            for line in file:
                # تقسيم السطر إلى اسم الفندق وعدد الغرف
                parts = line.split()
                if len(parts) == 2:
                    name = parts[0]
                    rooms = int(parts[1])
                    hotel_dict[name] = rooms
        return hotel_dict
    except FileNotFoundError:
        print("خطأ: الملف غير موجود!")
        return {}

# 2. دالة معالجة الحجز أو الإلغاء
def process_reservation(hotel_dict):
    if not hotel_dict:
        print("لا توجد بيانات فنادق للمعالجة.")
        return

    # طلب المدخلات من المستخدم
    action = input("ماذا تريد أن تفعل؟ (book / cancel): ").strip().lower()
    hotel_name = input("أدخل اسم الفندق: ").strip()

    # التأكد من وجود الفندق في القائمة
    if hotel_name in hotel_dict:
        if action == "book":
            if hotel_dict[hotel_name] > 0:
                hotel_dict[hotel_name] -= 1
                print(f"تم الحجز بنجاح في {hotel_name}. الغرف المتبقية: {hotel_dict[hotel_name]}")
            else:
                print(f"نعتذر، لا توجد غرف متاحة في {hotel_name}.")
        
        elif action == "cancel":
            hotel_dict[hotel_name] += 1
            print(f"تم إلغاء الحجز. الغرف المتاحة الآن في {hotel_name}: {hotel_dict[hotel_name]}")
        
        else:
            print("أمر غير معروف، يرجى اختيار 'book' أو 'cancel'.")
    else:
        print("عذراً، هذا الفندق غير موجود في نظامنا.")

# --- تشغيل البرنامج ---
# تحميل البيانات
data = load_hotel_data("hotels.txt")

# تنفيذ عملية حجز (يمكنك تكرارها حسب الحاجة)
if data:
    print("البيانات الحالية:", data)
    process_reservation(data)
    print("البيانات بعد التعديل:", data)
```

```python
#سوال العدد الاولى بتاع المحاضره 
def is_prime(number):
    if number<2:
        return False
    for i in range(2,int(number**.5)+1):
        if number%i==0:
            return False
    return True
    #تخزين البيانات فى ملف دون مسح البيانات القديمه
def export_primes_to_file(number):
    with open("input,txt","a")as f:
        f.write(str(number))
    print(f"تم اضافه الرقم {number}الى الملف بنجاح")
n=int(input("enter a number"))
if is_prime(n):
    export_primes_to_file(n)
else:
    print("the number isnot prime")
    
```

```python
#Write a program to read text and return a dictionary of word frequencies
text = "apple banana apple cherry banana apple"
words = text.split()
frequency = {}

for w in words:
    # إذا كانت الكلمة موجودة زد عدادها، وإلا ابدأ من 0
    frequency[w] = frequency.get(w, 0) + 1

print(frequency)
```

```python
#write a program to calculate max, min and average 
def get_statue(data):
    total=0
    max=data[0]
    min=data[0]
    for i in data:
        total+=i
        if i > max:max=i
        if i < min:min=i
    return {"max":max , "min":min , "aver":total/len(data)}
print(get_statue([3,4,7,8,9]))
            
```
