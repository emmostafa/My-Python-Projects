```python
#ادخال معاملات الخط المستقيم الاول نفرض مثلا a,b,c
a=float(input("enter a"))
b=float(input("enter b"))
c=float(input("enter c"))
#معاملات الخط المستقيم التانى نفرض مثلا E,m,n
E=float(input("enter E"))
v=float(input("enter m"))
n=float(input("enter n"))
#نجيب ميل كل خط مستقيم
m1=-a/b
m2=-E/v
#نشوف بقا هل المسيمان متعامده ام متوازيه ام متقاطعه ام متطابقه
if (m1*m2)==-1:
    print("الخطان متعامدان")
elif(m1==m2):#هنا فيه حالتين اما متوازيين اما متطابقين على حسب الجزء المقطوع لومتساوى يكون متطابقين غيز كدا متوزايين
    if (a/E == b/v ==c/n):
        print("متطابقان")
    else:
        print("متوازيان")
else:
    print("متقاطعان")
    
```
