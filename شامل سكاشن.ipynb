```python
students=[]
def add_student():
    name=input("enter the name")
    grade=int(input("enter the grade"))
    dic={"name":name,"grade":grade}
    students.append(dic)
def show_student():
    for v in students:
        print(v["name"],v["grade"])
def cal_aver():
    total=0
    for v in students:
        total+=v["grade"]
        ave=total/len(students)
        print("avarage",ave)
def highest_grade():
    max1=students[0]
    for v in students:
        if v["grade"]>max1["grade"]:
            max1=v
        print(max1["name"],max1["grade"])
while True:
    print("1 add student")
    print("2 show students")
    print("3 average")
    print("4 max1 student")
    print("5 exit")
    choice=input("chosse:")
    if choice=="1":
        add_student()
    elif choice=="2":
        show_student()
    elif choice=="3":
        cal_aver()
    elif choice=="4":
        highest_grade
    elif choice=="5":
        break
        
    
        
```
